```python
from sklearn.cluster import KMeans
list1 = []
a = iris[['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width',]].to_numpy()

for i in range(1,11):
    kmeans = KMeans(n_clusters =i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    
    kmeans.fit(a)
    list1.append(kmeans.inertia_)
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter =300, n_init = 10, random_state = 0)

y_kmeans = kmeans.fit_predict(a)
```

```python
plt.scatter(a[y_kmeans == 0, 0], a[y_kmeans == 0, 1], s = 100, c = 'paleturquoise', label = 'Iris-setosa')

plt.scatter(a[y_kmeans == 1, 0], a[y_kmeans == 1, 1], s = 100, c = 'plum', label = 'Iris-versicolour')

plt.scatter(a[y_kmeans == 2, 0], a[y_kmeans == 2, 1], s = 100, c = 'lightcoral', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')


plt.legend()
```
